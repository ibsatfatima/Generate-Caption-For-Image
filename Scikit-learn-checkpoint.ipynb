#### It's best to fill and transform training and test sets separately (as shown in the code below).

      The main takeaways:

    1: Split your data first (into train/test)
    2: Fill/transform the training set and test sets separately

```python
x = missing_data.drop('Price', axis = 1)
y = missing_data['Price']

#split the data
np.random.seed(42)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
# Now we need to work with x data means input or features data to remove missing data and transformation
x.isnull().sum()
```

```python
# Lets fill the missing values with sklearn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# Now fill catagorical values with missing and numerical with their mean
catagorical = SimpleImputer(strategy = 'constant', fill_value = 'missing')
doors = SimpleImputer(strategy = 'constant', fill_value = 4)
numerical = SimpleImputer(strategy = 'mean')

#Define columns 
cata_feature = ['Make', 'Colour']
door_feature = ['Doors']
num_feature = ['Odometer (KM)']

#Create an imputer 
imputation = ColumnTransformer([
    ('catagorical', catagorical, cata_feature),
    ('doors', doors, door_feature),
    ('numerical', numerical, num_feature)])

# Now fill the train and test values seperately
filled_x_train = imputation.fit_transform(x_train)
filled_x_test = imputation.fit_transform(x_test)
```

```python
filled_x_train
```

```python
# Now get transformed data into dataframes
train_x = pd.DataFrame(filled_x_train, columns = ['Make', 'Colour', 'Doors', 'Odometer (KM)'])
test_x = pd.DataFrame(filled_x_test, columns = ['Make', 'Colour', 'Doors', 'Odometer (KM)'])

# Check missing data in training set
train_x.isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

catagorical_features = ['Make', 'Colour', 'Doors']
encoder = OneHotEncoder()
transformer = ColumnTransformer([('encoder', encoder, catagorical_features)], remainder = 'passthrough')

# Fill the train and test data seperately
transformed_x_train = transformer.fit_transform(train_x)
transformed_x_test = transformer.fit_transform(test_x)

transformed_x_train.toarray()
```

```python
# Now we've transformed X, let's see if we can fit a model
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()

model.fit(transformed_x_train, y_train)
model.score(transformed_x_test, y_test)
```
